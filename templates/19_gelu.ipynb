# 🟢 Easy: GELU Activation

Implement the **GELU** (Gaussian Error Linear Unit) activation.

$$\text{GELU}(x) = x \cdot \Phi(x) = x \cdot 0.5 \cdot (1 + \text{erf}(x / \sqrt{2}))$$

### Signature
```python
def my_gelu(x: Tensor) -> Tensor: ...
```

### Rules
- Do NOT use `F.gelu`, `nn.GELU`, or `torch.nn.functional.gelu`
- Use `torch.erf` for the exact version

````python
import torch
import math
````

````python
'''

GELU: x * 一个概率
概率是： 0.5 * (1 + erf(x / sqrt(2))) 这是标准正态分布的CDF

| x 值  | GELU 行为     |
| ---- | ----------- |
| 很负   | ≈ 0（被压掉）    |
| 0 附近 | 保留一部分       |
| 很正   | ≈ x（基本原样通过） |

--> 平滑版本的relu

'''

'''
数学部分：本质用正太分布的概率来决定x保留多少

Φ(x) 没有简单初等函数表达式， 用上面的公式进行替代。 意思是👉 “一个标准正态变量 ≤ x 的概率”
torch.erf(z)： 比较类似于一个-1，1 的s型函数， 也没有解析解 是近似计算的


√2 是为了把正态分布的指数变成 erf 的标准形式
标准正太分布的pdf： 1/sqrt(2pi) * exp(-x^2/2), 指数是 exp(-x^2/2)
erf是 exp(-x^2) 需要换一下变量

0.5 * (1 + erf(...)) -> (-1, 1) → (0, 1) 变成了概率权重
'''

def my_gelu(x):
    return x * 0.5 * (1 + torch.erf( x / math.sqrt(2)))
````

````python
# 🧪 Debug
x = torch.tensor([-2., -1., 0., 1., 2.])
print('Output:', my_gelu(x))
print('Ref:   ', torch.nn.functional.gelu(x))
````
