# 🟡 Medium: Simple Linear Layer

Implement a fully-connected linear layer: **y = xW^T + b**

### Signature
```python
class SimpleLinear:
    def __init__(self, in_features: int, out_features: int): ...
    def forward(self, x: torch.Tensor) -> torch.Tensor: ...
```

### Requirements
- `self.weight`: shape `(out_features, in_features)`, init with `randn * (1/√in_features)`
- `self.bias`: shape `(out_features,)`, init as zeros
- Both must have `requires_grad=True`
- `forward(x)` computes `x @ W^T + b`
- Do **NOT** use `torch.nn.Linear`

````python
import torch
import math
````

````python
# 只是为了实现y = x * W.T + b
# 这里的 w 是（out_feature, in_feature）

class SimpleLinear:
    def __init__(self, in_features: int, out_features: int):
        self.in_features = in_features
        self.out_features = out_features

        # X: (N, in_features)
        # W : (out_features, in_features)
        # out_featuers 是神经元的个数 每一个输出维度是一个神经元
        # 每个神经元有一组权重

        # Xavier / variance scaling 初始化
        # 如果不缩小 输入维度越大 输出的 variance就会越来越大
        scale = 1.0 / (in_features ** 0.5)
        self.weight = torch.randn(out_features, in_features) * scale
        self.weight.requires_grad_() # 让 torch 记住梯度

        #
        self.bias = torch.zeros(out_features)
        self.bias.requires_grad_() # 让 torch 记住梯度

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return x @ self.weight.T + self.bias
````

````python
# 🧪 Debug
layer = SimpleLinear(8, 4)
print("W shape:", layer.weight.shape)  # should be (4, 8)
print("b shape:", layer.bias.shape)    # should be (4,)

x = torch.randn(2, 8)
y = layer.forward(x)
print("Output shape:", y.shape)        # should be (2, 4)
````
