# 🟢 Easy: Implement Softmax

Implement the **Softmax** function from scratch.

$$\text{softmax}(x_i) = \frac{e^{x_i}}{\sum_j e^{x_j}}$$

### Signature
```python
def my_softmax(x: torch.Tensor, dim: int = -1) -> torch.Tensor:
    ...
```

### Rules
- Do **NOT** use `torch.softmax`, `F.softmax`, or `torch.nn.Softmax`
- Must be **numerically stable** (hint: subtract `max` before `exp`)

### Example
```
Input:  tensor([1., 2., 3.])
Output: tensor([0.0900, 0.2447, 0.6652])  # sums to 1.0
```

torch.sum(x_exp) sums all elements in the tensor into a single scalar.
to make it stable, need to substract MAX first, to cap the values as exp(0)  = 1. 解决上溢问题 overflow


对很小的值 减去最大值仍然能保证最大的数调整后的值是0， 所以exp（最大的数）还是1

为什么softmax叫softmax？

它是一个“软化”的 argmax 函数：它不是直接给出最大值的位置，而是给出每个位置是最大值的“软概率”

这种方式让它在数学上是连续可微分的，这在训练神经网络时非常重要，因为梯度下降等优化算法需要函数的梯度信息。


````python
def my_softmax(x: torch.Tensor, dim: int = -1) -> torch.Tensor:
    x_max = torch.max(x, dim=dim, keepdim=True).values
    x_exp = torch.exp(x-x_max)
    x_sum = torch.sum(x_exp, dim=dim, keepdim=True)
    return x_exp / x_sum


````

````python
# 🧪 Debug
x = torch.tensor([1.0, 2.0, 3.0])
print("Output:", my_softmax(x, dim=-1))
print("Sum:   ", my_softmax(x, dim=-1).sum())  # should be ~1.0
print("Ref:   ", torch.softmax(x, dim=-1))
````
