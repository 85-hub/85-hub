# рҹҹў Easy: Implement Dropout

Implement **Dropout** regularization from scratch.

### Signature
```python
class MyDropout(nn.Module):
    def __init__(self, p: float = 0.5): ...
    def forward(self, x: Tensor) -> Tensor: ...

### Rules
- During **training**: zero each element with probability `p`, scale remaining by `1/(1-p)`
- During **eval**: return input unchanged (identity)
- Do NOT use `nn.Dropout` or `F.dropout`

````python
import torch
import torch.nn as nn
````

````python
# вңҸпёҸ YOUR IMPLEMENTATION HERE

class MyDropout(nn.Module): # nn.ModuleжҳҜжүҖжңүзҪ‘з»ңеұӮзҡ„еҹәзұ»
# иҮӘе·ұеҶҷдёҖеұӮзҡ„ж—¶еҖҷпјҢйҖҡеёёйғҪз»§жүҝе®ғ
    def __init__(self, p=0.5):
      # еӣ дёәдҪ з»§жүҝдәҶ nn.ModuleпјҢжүҖд»ҘиҰҒе…ҲжҠҠзҲ¶зұ»йӮЈдёҖеҘ—еҲқе§ӢеҢ–еҘҪгҖӮеҗҰеҲҷ PyTorch зҡ„еҫҲеӨҡжңәеҲ¶дёҚиғҪжӯЈеёёе·ҘдҪңгҖӮ
        super().__init__()
        self.p = p
        # p = 1 д№ҹдёҚиЎҢпјҢеӣ дёәеҗҺйқўиҰҒйҷӨд»Ҙ 1-pпјҢдјҡеҸҳжҲҗйҷӨд»Ҙ 0
        if not 0 <= self.p <= 1:
            raise ValueError


    def forward(self, x):
      # dropout зҡ„и§„еҲҷжҳҜдёҚз®ЎеҪўзҠ¶пјҢзӣҙжҺҘеҜ№жҜҸдёӘе…ғзҙ зӢ¬з«ӢйҡҸжңәеӨ„зҗҶгҖӮ

      # иҝҷжҳҜ dropout зҡ„ж ёеҝғи§„еҲҷд№ӢдёҖпјҡ
      # и®ӯз»ғж—¶пјҡйҡҸжңәдёўдёҖдәӣе…ғзҙ 
      # жөӢиҜ•/жҺЁзҗҶж—¶пјҡдёҚиғҪеҶҚйҡҸжңәдёўдәҶпјҢзӣҙжҺҘеҺҹж ·иҝ”еӣһ

      # self.training жҳҜ nn.Module иҮӘеёҰзҡ„дёҖдёӘеёғе°”еҖјпјҡ

      # model.train() еҗҺпјҢself.training == True
      # model.eval() еҗҺпјҢself.training == False
      if self.training == False:
        return x

      keep_prob = 1 - self.p
      # torch.rand_like(x) з”ҹжҲҗдёҖдёӘе’Ң x еҪўзҠ¶дёҖжЁЎдёҖж ·зҡ„еј йҮҸпјҢйҮҢйқўжҜҸдёӘе…ғзҙ йғҪжҳҜ [0, 1) д№Ӣй—ҙзҡ„йҡҸжңәж•°гҖӮ
      # жҠҠе®ғиҪ¬жҲҗе’Ң x еҗҢж ·зҡ„ж•°жҚ®зұ»еһӢ
      mask = (torch.rand_like(x) < keep_prob).to(x.dtype)

      # inverted dropout
      return x * mask / keep_prob



````

````python
# рҹ§Ә Debug
d = MyDropout(p=0.5)
d.train()
x = torch.ones(10)
print('Train:', d(x))
d.eval()
print('Eval: ', d(x))
````
